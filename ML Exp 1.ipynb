```python
#from google.colab import files
#uploaded=files.upload()

#24BAD128-VIBUSHA-SCENARIO 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path=r"data.csv"
data=pd.read_csv(path,encoding="latin1")

print("24BAD128-VIBUSHA-SCENARIO 1")
print(data.head())
print(data.tail())
print(data.info())
print(data.describe())
print(data.isnull()) #Checks missing value
print(data.columns)

#Bar Chart
country_counts = data["Country"].value_counts().head(10)

plt.figure()
plt.bar(country_counts.index, country_counts.values)
plt.xlabel("Country")
plt.ylabel("Number of Transactions")
plt.title("Top 10 Countries by Number of Transactions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Line Chart
data["Sales"] = data["Quantity"] * data["UnitPrice"]
data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"])

daily_sales = data.groupby(data["InvoiceDate"].dt.date)["Sales"].sum()

plt.figure()
plt.plot(daily_sales.index, daily_sales.values)
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.title("Daily Sales Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#from google.colab import files
#uploaded=files.upload()

#24BAD128-VIBUSHA.G-SCENARIO 4
import pandas as pd
import matplotlib.pyplot as plt

path = r"marketing_campaign.csv"
data = pd.read_csv(path, encoding="latin1", sep="\t")

print("24BAD128-VIBUSHA.G-SCENARIO 4")
print("\nFirst 5 rows:\n", data.head())
print("\nDataset Info:\n")
print(data.info())

print("\nMissing values in each column:\n")
print(data.isnull().sum())

data.columns = data.columns.str.strip()
data["Age"] = 2026 - data["Year_Birth"]

plt.hist(data["Age"], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

plt.boxplot(data["Age"], vert=False)
plt.title("Age Boxplot")
plt.show()

plt.hist(data["Income"].dropna(), bins=20)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()

plt.boxplot(data["Income"].dropna(), vert=False)
plt.title("Income Boxplot")
plt.show()

spend_cols = ["MntWines","MntFruits","MntMeatProducts",
              "MntFishProducts","MntSweetProducts","MntGoldProds"]

data["TotalSpend"] = data[spend_cols].sum(axis=1)

plt.hist(data["TotalSpend"], bins=20)
plt.title("Total Spending Distribution")
plt.xlabel("Total Spend")
plt.ylabel("Count")
plt.show()

plt.boxplot(data["TotalSpend"], vert=False)
plt.title("Total Spending Boxplot")
plt.show()
```
